```python
import pandas as pd

#make a pivot table that says how much male and female spend in each catagory
df_sales = pd.read_excel('supermarket_sales.xlsx')  # Changed from pd.read_csv() to pd.read_excel() for .xlsx files

# Make a pivot table that says how much male and female spend in each product line
df_sales.pivot_table(index='Gender', columns='Product line',
                     values=['Total'], aggfunc='sum')
```

# making a Pivot Table
.pivot(): Returns reshaped Dataframe organized by give index/column values('pivot without aggregation')


```python
import pandas as pd 
df_pop = pd.read_csv('population_total.csv')
df_pop
```

```python
#dropping null values
df_pop.dropna(inplace=True) 
```

```python
#making a pivot table
df_pivot = df_pop.pivot(index='year', columns='country', values='population')
```

```python
#selecting some countries
df_pivot = df_pivot[['United States', 'India', 'China', 'Indonesia', 'Brazil']] #overwriting the dataframe to the same varriable
df_pivot
```
